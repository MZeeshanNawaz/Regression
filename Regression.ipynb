```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error, r2_score

data = {
    'Day': ['sunny','windy','sunny','windy','windy'],
    'Temp': [20,18,16,28,32],
    'Score': [100,70,50,30,25]
}
df = pd.DataFrame(data)
onehot_encoder = OneHotEncoder()
encoded_day = onehot_encoder.fit_transform(df[['Day']]).toarray()
x = np.concatenate([encoded_day,df[['Temp']]],axis=1)
y = df['Score']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)

model = MLPRegressor()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)

mse = print("Mean sqaure error ", mean_squared_error(y_test,y_pred))
r2 = print("R2 Score", r2_score(y_test,y_pred))

new_instance = pd.DataFrame([['sunny',15]],columns = ['Day','Temp'])
encoded_new_day = onehot_encoder.transform(new_instance[['Day']]).toarray()
new_instance = np.concatenate([encoded_new_day,new_instance[['Temp']]],axis=1)
predicted_score = model.predict(new_instance)
print("Predicted Score:", round(predicted_score[0], 2))
```
